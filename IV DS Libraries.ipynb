More Series Methods:
- `.any()`: returns single boolean. Do any values in the series meet the condition?
- `.all()`: returns single boolean. Do all values in the series meet the condition?
- `.isin()`: Returns series of boolean values. Is each string in your series in the list of strings? 
- `.apply()`: apply a function to each item in a series.
- `.replace()`: replace/rename values (typically categorical values)
    - e.g.: `df["EDU"] = df.EDU.replace({1: "<9", 2: "9-11", 3: "HS/GED", 4: "Some college/AA", 5: "College", 9: "Don't know"})`
- `.fillna('new_val'/0)` replace nulls with value
- `.dropna()`

###### Part 2

*Indexing and Subsetting*

- can pass booling series to call indices like `df[<bool Series>]`
- `.loc` select specific rows AND columns by index **label**, inclusive to end label
    - e.g. `df.loc[row_idxs, col_lbls]`
    - `row_idxs` can be index labels or bool series
- `df.iloc[row_idxs, col_idxs]` select specific rows AND columns by index **position**. non-inclusive to end position

*Aggregating*

**`agg`**

`df.col.agg('<agg_funct(s)>')`

`df[[<cols>]].agg('<agg_funct(s)>')`

**`groupby`**

`df.groupby('col_cat(s)').col_cont.<func>()`

`df.groupby('col_cat(s)').col_cont.agg(['min', 'mean', 'max'])`

can add describe:

`df.groupby('col').col2.describe()`

**`np.where`** (Like `IF()` in SQL)

`df['new_col'] = np.where((df.col conditional), val_if_true, val_if_false)`

**`transform`** (Like CASE statements in SQL) produces a series with the same length of the original df where each value represents the aggregation from the subgroup resulting from the .groupby

`df.assign(new_col = df.groupby('col').math.transform('<func>'))`

*Merging and Joining*

**`pd.concat`**

`pd.concat([df1, df2], axis=0, join='outer', ignore_index=True)` 0 for row-wise (think "more rows added"), 1 for column-wise

**`.merge`** (Like `JOIN` in SQL)

[SQL comparison](https://pandas.pydata.org/pandas-docs/stable/getting_started/comparison/comparison_with_sql.html#compare-with-sql-join) to JOIN

`dfL.merge(dfR, left_on='col1', right_on='col2', how='jointype')`

###### Part 3

*Reshaping*

**`pd.crosstab`** pivot/contingency table

`pd.crosstab(df.col_cat, df.col_cat2, margins=True)` 

`pd.crosstab(df.col_cat, df.col_cat2, normalize=True).round(3)` 

- normalize row-wise: `.apply(lambda z: z/z.sum(), axis=0)`
- normalize column-wise: `.apply(lambda z: z/z.sum(), axis=1)`

**`.pivot_table`** pivot table

`df.pivot_table(index='col_cat', columns='col_cat2', values='col_cont', aggfunc=<func(s)>)`

**`.map`**

can use a dictionary to save new info (e.g. calculations) to a new column. ex. below

````python
n = 40
orders = pd.DataFrame({
    'drink': np.random.choice(['Tea', 'Water', 'Water'], n),
    'meal': np.random.choice(['Curry', 'Yakisoba Noodle', 'Pad Thai'], n),
})
prices = {
    'Yakisoba Noodle': 9,
    'Curry': 11,
    'Pad Thai': 10,
    'Tea': 2,
    'Water': 0,
}
orders['bill'] = orders.drink.map(prices) + orders.meal.map(prices)

orders.head()
````

````python
pd.crosstab(orders.drink, orders.meal, normalize=True, margins=True)
````

````python
pd.crosstab(orders.drink, orders.meal, normalize=True)
````

````python
# avg bill
orders.pivot_table(index='drink', columns='meal', values='bill')
````

````python
orders.groupby(['drink', 'meal']).bill.mean()
````

*subplots*

- `plt.figure(figsize=(w, h))`
- `plt.subplots(n_rows, n_cols, plot_pos)`
- `plt.subplots(132)` where 1 = `n_rows`, 3 = `n_cols`, 2 = `plot_pos`
- `plt.suptitle('Subplots Demo')`
- `plt.subplots_adjust(hspace=0.5, wspace=0.5)` to add space between subplots

````python
n_rows = 1
n_cols = 2

x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 2, 3, 4, 5]

plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y)
plt.title('x ~ y')

plt.subplot(n_rows, n_cols, 2)
plt.plot(x, z)
plt.title('x ~ z')

plt.show()
````

````python
# Create a figure with 2 subplots
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot([1, 2, 3], [4, 5, 6])
ax1.set_title('Subplot 1')
ax2.plot([1, 2, 3], [6, 5, 4])
ax2.set_title('Subplot 2')
````

### [seaborn](https://seaborn.pydata.org/api.html)

*Scatter charts*

- `sns.relplot(x='x_col', y='y_col', col='group_col', hue='group_col', style='group_col', data=df)`
    - `col` creates subplots of groups
    - `hue` changes data colors by group
    - `style` changes data markers by group
    - `.set_title('title')`
    - `.set_ylabel('ylabel')`
    - `.set(title='', ...)`
-- `sns.regplot(x='x_col', y='y_col', data=df,  fit_reg=False, scatter_kws={"alpha": 0.2})`

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

mpg = sns.load_dataset('mpg')
sns.relplot(x='horsepower', y='mpg', hue='origin', data=mpg)
plt.title('Miles per Gallon vs Horsepower')
plt.xlabel('Horsepower')
plt.ylabel('Miles Per Gallon')
````

*Other Charts*

Histogam:
- `kde` adds approx distribution for continuous variables

````python
sns.histplot(mpg.horsepower, kde = True)
````

Box plots

- `sns.boxplot(data=mpg, y='cont_col', x='cat_col', hue='cat_col2')`

````python
plt.figure(figsize=(12, 10))
plt.suptitle('Boxplots with Seaborn')

plt.subplot(221)
sns.boxplot(data=mpg, y='mpg')
plt.title('A "typical" boxplot')

plt.subplot(222)
sns.boxplot(data=mpg, x='mpg')
plt.title('Rotated 90$\degree$')

plt.subplot(223)
sns.boxplot(data=mpg, y='mpg', x='origin')
plt.title('Providing an x and y value groups the data')

plt.subplot(224)
sns.boxplot(data=mpg, y='mpg', x='origin', hue='cylinders')
plt.title('You can also use other kwargs, like hue')

plt.subplots_adjust(hspace=0.4)
````
